```python
j = []
for i in text_list :
    sentences = hannanum.nouns(i)
    result = []
    #print('sentences')
    #print(sentences)
    for word in sentences :
        #print(word)
        if len(word) > 1 :
            result.append(word)
    #print('result')
    #print(result)
    j.append(result)

print(len(j))
print(j, '\n')
```

```python
bigram = gensim.models.Phrases(j)
trigram = gensim.models.Phrases(bigram[j])
bigram_model = gensim.models.phrases.Phraser(bigram)
gtrigram_model = gensim.models.phrases.Phraser(trigram)
```

```python
bigram_document = [bigram_model[nouns] for nouns in j]
print(bigram_document[1])
```

```python
test_id2word = corpora.Dictionary(bigram_document)
test_corpus = [test_id2word.doc2bow(doc) for doc in bigram_document]
print(test_corpus[0])
```

```python
test_coherence_score=[]
for i in range(2,10):
    model = gensim.models.ldamodel.LdaModel(corpus=test_corpus, id2word=test_id2word, num_topics=i)
    coherence_model = CoherenceModel(model, texts=bigram_document, dictionary=test_id2word, coherence='c_v')
    coherence_lda = coherence_model.get_coherence()
    print('n=',i,'\nCoherence Score: ', coherence_lda)
    test_coherence_score.append(coherence_lda)
```
